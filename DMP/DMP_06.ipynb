# Лабораторна робота №6 з Видобування даних

## Тема. Етап розгортання моделі згідно методології CRISP-DM.

### Мета роботи: навчитись використовувати натреновані моделі в клієнтських застосунках та візуалізовувати результати дослідження.

#### Для отримання максимальної оцінки потрібно виконати розгортання моделей для двох задач, які виконувались в попередніх роботах.

### 1. Опишіть варіанти використання ваших моделей в реальному житті та приклади даних, які потрібні для їх застосування.

### 2. Виконайте серіалізацію натренованих моделей. Для збережння стану натренованої моделі ми використаємо [pickle](https://scikit-learn.org/stable/modules/model_persistence.html) - стандартний інструмент для серіалізації у python.

### 3. Розробіть REST API для використання навчених моделей. Збережену на попередньому кроці модель потрібно завантажити і використати у вашому коді.

Якщо ви виконували кластеризацію і не бачите користі у її використанні в якості моделі прогнозування, то можете виконати візуалізацію (крок 5) замість написання REST API для цієї задачі.

### 4. Продемонструйте роботу ваших моделей в Postman або простому веб-інтерфейсі.

### 5. В контексті exploratory data analysis та нових знахідок, що були отримані під час моделювання (зокрема, за результатами кластеризації та опису отриманих груп даних), зробіть одну візуалізацію.

Обгрунтуйте вибраний [тип графіку](https://www.data-to-viz.com/) для виших даних і інформацію, яку ви хочете донести читачу.

Старайтесь зробити візуалізацію даних і ваших знахідок з урахуванням [рекомендацій](https://stephanieevergreen.com/data-visualization-checklist/).
